```python
def mse(yhat,y):
    err = 0
    for i in range(len(y)):
        err = err + (yhat[i]-y[i])**2
    return err/len(y)
```

```python
def acc_(predictions,testSet_y):
  test_size = len(testSet_y)
  # testSet_y = list(testSet_y)
  correct_predictions = 0
  for i in range(test_size):
    if testSet_y[i]==predictions[i]:
      correct_predictions= correct_predictions+1
  return (correct_predictions/test_size)*100
```

```python
def normalise_(data_):
    desc = data_.describe()
    for col in desc.columns.unique():
        xmean = desc[col][1]
        sigma = desc[col][2]
        data_[col] = (data_[col]-xmean)/sigma
    return data_
```

```python
def kfoldsplit(k,data,target):
    
    splits = []
    
    y = pd.DataFrame(data[target])
    x = data.drop(target,axis=1)
    d = []
    s = int(len(data)/k)
    
    for i in range(k):
        a = s*i
        b = a+s
        d.append([a,b])
    for [a,b] in d:
        
        
        x_test = x[a:b]
        y_test = np.array(y[a:b][target])
        x_train = pd.concat([x[:a],x[b:]])
        y_train = np.array(pd.concat([y[:a],y[b:]])[target])
        splits.append([x_train,y_train,x_test,y_test])
    
    return splits
```

```python
def _sigmoid_(arg):
    return (1/(1+np.exp(-arg)))
```

```python
def loss(yreal,ypred):
    return np.mean(-yreal*np.log(ypred)-(1-yreal)*np.log(1-ypred)) 
```

```python
class LogRegression(object):
    """docstring for LogRegression."""
    def __init__(self, arg):
        super(LogRegression, self).__init__()
        self.arg = arg
        
    def fit(self,arg):
        
        xtrain,ytrain,xtest,ytest,no_epochs,alpha,l2 = arg
        n = xtrain.shape[1]
        d = xtrain.shape[0]
        theta = np.zeros((d,1))
        bias = 0
        lossestr = []
        lossests = []
        testacc = []
        trainacc = []

        for e in range(no_epochs):
            #print(f"epoch: {e+1}")
            # print(( np.dot(x, np.array(theta).reshape(d,1))).shape)
           
            yhat=_sigmoid_( np.dot(xtrain.T,theta)+ bias)
            ythat=_sigmoid_(np.dot(xtest.T,theta)+ bias)
            yhat = yhat.reshape(yhat.shape[0])
            ythat = ythat.reshape(ythat.shape[0])
            # print(acc_(yhat,ytrain))
            # print(acc_(ythat,ytest))
            

            G,db= gradient(xtrain,ytrain,yhat,l2,theta)
            
            theta -= alpha*G.reshape((d,1))
            self.weights = theta
            bias= (bias - alpha*db)
            # print(bias.shape)
            self.B = bias

            trainacc.append(acc_(self.predict(xtrain),ytrain))
            testacc.append(acc_(self.predict(xtest),ytest))

            lossestr.append(loss(ytrain,yhat))
            lossests.append(loss(ytest,ythat))
        
        self.lossestr = lossestr
        self.lossests = lossests
        self.trainingaccuracies = trainacc
        self.testaccuracies = testacc

    def predict(self,X_test):
        # d = len(X_test.columns.unique())
        p = _sigmoid_( np.dot(X_test.T,self.weights)+ self.B)
        predictedValues = []
        for i in p:
            if i[0]>0.5:
                predictedValues.append(1)
            else:
                predictedValues.append(0)
        y_predicted = predictedValues
        return y_predicted
    def get_mse(self,xtest,ytest):
        p = _sigmoid_( np.dot(xtest.T,self.weights)+ self.B)
        return mse(p,ytest)


def gradient(xtrain,ytrain,yhat,l2,theta):
    
    theta = theta.reshape(theta.shape[0])
    s = 0
    for i in range(len(xtrain.T)):
        s+=((yhat[i]-ytrain[i])*xtrain.T[i])
    cost =  s
    reg = theta*l2
    delTheta =  (cost+reg)
    delTheta = delTheta/len(xtrain.T)
    delBias = np.sum(yhat-ytrain)
    delBias = delBias/len(xtrain.T)
    # print(f"delbias={delBias.shape}")
    #print(delTheta.shape)
    return delTheta,delBias
```

```python
def fitgrid(l2values,k,target,estimator,savemodel=False,getplots=False):
   maxAccuracy = float('-inf')
   l2best = 0
   L2ACC = []
   S = kfoldsplit(k,df,target)
   for l2 in l2values:
      for i in range(len(S)):
         xtrain,ytrain,xtest,ytest = S[i]
         xtrain = normalise_(xtrain)
         xtest = normalise_(xtest)

         if estimator == 'implemented':
            lr = LogRegression(None)
            xtrain = np.array(xtrain).T
            xtest = np.array(xtest).T
            c = lr.fit([xtrain,ytrain,xtest,ytest,500,0.1,l2])
            ytestp = lr.predict(xtest)
            ytrainp = lr.predict(xtrain)
            L_train = lr.lossestr
            L_test = lr.lossests
            if getplots == True:
               plt.title(f"l2 = {l2}")
               sns.lineplot(x=range(len(L_test)),y=L_test,label=f"fold {i+1} test losses")
               sns.lineplot(x=range(len(L_train)),y=L_train,label=f"fold {i+1} train losses")
               plt.show()
               tsa = lr.testaccuracies
               tra = lr.trainingaccuracies
               plt.title(f"accuracies at l2 = {l2}")
               sns.lineplot(x=range(len(tsa)),y=tsa,label='test accuracy')
               sns.lineplot(x=range(len(tra)),y=tra,label ='train accuracy')
               plt.show()
            if savemodel==True:
               joblib.dump(c,f"savedmodels/q2fold{i}.pkl")
            
            L2ACC.append([i+1,float(l2),acc_(ytestp,ytest),acc_(ytrainp,ytrain),
            lr.get_mse(xtest,ytest)[0],
            lr.get_mse(xtest,ytest)[0]])

         elif estimator == 'sklearnLogisticRegression':
            clf = LogisticRegression(random_state=0,penalty='l2').fit(xtrain, ytrain)
            ytestp = clf.predict(xtest)
            ytrainp = clf.predict(xtrain)
            L2ACC.append([i+1,acc_(ytestp,ytest),acc_(ytrainp,ytrain),
                           mse(clf.predict_proba(xtest).T[1],ytest),
                           mse(clf.predict_proba(xtrain).T[1],ytrain)]
                           )
            if savemodel==True:
               joblib.dump(clf,f"savedmodels/q2fold{i}sklearn.pkl")

         else:
            print("please properly provide the estimator as either implemented or sklearnLogisticRegression")
            return

   if estimator == 'implemented':         
      L2ACC = pd.DataFrame(L2ACC,columns=['Fold','Lambda Value','Test Accuracy','Training Accuracy','Test Error','Train Error'])
      maxAccuracy = L2ACC[L2ACC['Test Accuracy']==L2ACC['Test Accuracy'].max()]['Test Accuracy'].iloc[0]
      l2best = L2ACC[L2ACC['Test Accuracy']==L2ACC['Test Accuracy'].max()]['Lambda Value'].unique()
      return L2ACC,maxAccuracy,l2best
   elif estimator == 'sklearnLogisticRegression':
      L2ACC = pd.DataFrame(L2ACC,columns=['Fold','Test Accuracy','Training Accuracy','Test Error','Train Error'])
      maxAccuracy = L2ACC[L2ACC['Test Accuracy']==L2ACC['Test Accuracy'].max()]['Test Accuracy'].iloc[0]
      return L2ACC,maxAccuracy     
```

```python
def gridsearch__(l2values,numberOfFolds,targetVariable):
    return fitgrid(l2values,numberOfFolds,targetVariable,estimator='implemented')
    
```

```python
l2values = [0,0.0001,0.0006,0.003,0.1,0.5,100,300,40000,float('inf')]
```

```python
RECORDS,MAXACC,BESTLambda = gridsearch__(l2values,5,'Outcome')
```

```python
print(f"maxaccuracy achieved: {MAXACC} best lambda values: {BESTLambda} mean accuracy: {RECORDS['Test Accuracy'].mean()}")
```
